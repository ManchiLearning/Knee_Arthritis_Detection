```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from load_dataset import load_dataset
import numpy as np
```

```python
imgs, labels, classes = load_dataset("dataset")
```

```python
X_train, X_test, y_train, y_test = train_test_split(imgs, labels, test_size=0.2, random_state=42)
```

```python
X_train_res = X_train.reshape((X_train.shape[0], X_train.shape[1] * X_train.shape[2]))
X_test_res = X_test.reshape((X_test.shape[0], X_test.shape[1] * X_test.shape[2]))
```

```python
X_train_res = X_train_res / 255
X_test_res = X_test_res / 255
```

```python
model = MLPClassifier((64,32),max_iter=10000)
model.fit(X_train_res, y_train)
```

```python
model.score(X_test_res, y_test)
```
